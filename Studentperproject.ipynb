```python
plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.set_palette('dark')
ax=sns.countplot(data=df,x='Gender',palette='bright')
plt.title('Distribution of gender')
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.set_palette('bright')
ax=sns.countplot(data=df, x='Race', palette='plasma')
plt.title('Distribution of race')
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.show()
```

```python
print("\nCross Tabulation between Gender and Race:")
print(pd.crosstab(df['Gender'], df['Race']))
```

```python
plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.set_palette('bright')
ax=sns.countplot(data=df, x='Race', palette='flare',hue='Gender')
plt.title('Distribution of race')
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.set_palette('muted')
ax=sns.histplot(data=df, x='Parents Education')
plt.title('Comparision of Parental Education')
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['Math score'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Math Scores')
plt.subplot(1, 3, 2)
sns.histplot(df['Reading score'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Reading Scores')
plt.subplot(1, 3, 3)
sns.histplot(df['Writing score'], bins=20, kde=True, color='limegreen')
plt.title('Distribution of Writing Scores')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
sns.boxplot(x='Gender', y='Math score', data=df, palette='Set1')
plt.title('Math Scores by Gender')
plt.subplot(1, 3, 2)
sns.boxplot(x='Gender', y='Reading score', data=df, palette='Set1')
plt.title('Reading Scores by Gender')
plt.subplot(1, 3, 3)
sns.boxplot(x='Gender', y='Writing score', data=df, palette='Set1')
plt.title('Writing Scores by Gender')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
sns.boxplot(x='Race', y='Math score', data=df, palette='BuPu')
plt.title('Math Scores by Race')
plt.subplot(1, 3, 2)
sns.boxplot(x='Race', y='Reading score', data=df, palette='GnBu')
plt.title('Reading Scores by Race')
plt.subplot(1, 3, 3)
sns.boxplot(x='Race', y='Writing score', data=df, palette='OrRd')
plt.title('Writing Scores by Race')
plt.tight_layout()
plt.show()
```

```python
average_scores = df.groupby('Race')[['Math score', 'Reading score', 'Writing score']].mean()
print(average_scores)
highest_avg_race = average_scores.mean(axis=1).idxmax()
highest_avg_scores = average_scores.loc[highest_avg_race]

print(f"The race/ethnicity group with the highest average test scores is '{highest_avg_race}'")
print("Average Scores:")
print(highest_avg_scores)
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1, 3, 1)
sns.boxplot(x='Parents Education',y='Math score',data=df, palette='Reds')
plt.title('Math Scores by Parental Education')
plt.xticks(rotation=0)

plt.subplot(1, 3, 2)
sns.boxplot(x='Parents Education',y='Writing score',data=df, palette='Greens')
plt.title('Writing Scores by Parental Education')
plt.xticks(rotation=0)

plt.subplot(1, 3, 3)
sns.boxplot(x='Parents Education',y='Reading score',data=df, palette='Blues')
plt.title('Reading Scores by Parental Education')
plt.xticks(rotation=0)


plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Lunch', y='Math score', data=df, ci=None, palette='bright')
plt.title('Average Math Scores by Lunch Type')
plt.show()
```

```python
df['Mean score'] = df[['Math score', 'Reading score', 'Writing score']].mean(axis=1)

plt.figure(figsize=(8, 6))
sns.barplot(x='Lunch', y='Mean score', data=df, palette='Set2')
plt.title('Mean Score by Lunch Type')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
sns.violinplot(x='Test preparation course', y='Math score', data=df, palette='RdPu',saturation=10)
plt.title('Math Scores by Test Preparation Course')
plt.subplot(1, 3, 2)

sns.violinplot(x='Test preparation course', y='Writing score', data=df, palette='RdPu',saturation=10)
plt.title('Writing scores by Test Preparation Course')
plt.subplot(1, 3, 3)

sns.violinplot(x='Test preparation course', y='Reading score', data=df, palette='RdPu',saturation=10)
plt.title('Reading Scores by Test Preparation Course')

plt.show()
```
